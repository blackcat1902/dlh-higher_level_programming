```python
#!/usr/bin/python3
import random
number = random.randint(-10000, 10000)
lastdigit = number % 10
if lastdigit > 5:
    print("Last digit of",number,"is",lastdigit,"and is greater than 5")
elif lastdigit == 0:
    print("Last digit of",number,"is",number % 10," and is 0")
elif lastdigit < 6 and number % 10 != 0:
    print("Last digit of",number,"is",number % 10,"and is less than 6 and not 0")
```
